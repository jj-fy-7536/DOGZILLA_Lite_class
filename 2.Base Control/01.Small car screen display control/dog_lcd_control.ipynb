```python
#导入xgoedu  #Import xguodu
from xgoedu import XGOEDU 
#实例化edu #Instantiate edu
XGO_edu = XGOEDU()
```

```python
#画直线 #Draw a straight line
XGO_edu.lcd_clear()
XGO_edu.lcd_line(80,80,150,150,color=(255,0,0),width=2)
```

```python
#画圆 #Draw a circle
XGO_edu.lcd_clear()
XGO_edu.lcd_round(30,30,10,color=(255,0,0),width=2)
```

```python
#画矩形 #Draw a rectangle
XGO_edu.lcd_clear()
XGO_edu.lcd_rectangle(20,20,100,100,fill=None,outline=(255,0,0),width=2)
```

```python
#显示文字 #Display Text
XGO_edu.lcd_clear()
XGO_edu.lcd_text(50,50,'hello',color=(255,0,0),fontsize=50)
```

```python
#显示图片 需要把图片放在/home/pi/xgoPictures 目录下，并且大小为320*240
#To display images, they need to be placed in the/home/pi/xgoPictures directory and have a size of 320 * 240
XGO_edu.lcd_picture("yahboom.jpg")
```

```python
#清除屏幕 #Clear screen
XGO_edu.lcd_clear()
```
